```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import numpy as np
import sys
import openpyxl
```

```python
df = pd.read_excel("Data for data visuliazation_v4.xlsx")
```

```python
unique_values = df['Medical Domain Category'].unique()
print(unique_values)
```

```python
df['Medical Domain Category'] = df['Medical Domain Category'].replace('Others', 'General Medicine, e.g., Internal Medicine')
df.to_excel("Data for data visuliazation_v4.xlsx", index=False) 
```

# Bar Charts and Picture of Countries For Medical Domain

United States

```python
# bar chart
df_filtered = df[df['Country_first author'] == 'United States']

category_counts = df_filtered['Medical Domain Category'].value_counts().reset_index()
category_counts.columns = ['Medical Domain Category', 'Count']

fig = px.bar(category_counts, 
             x='Count', 
             y='Medical Domain Category', 
             orientation='h', 
             title='<b>Medical Domain for United States</b>',
             labels={'Count':'Count', 'Medical Domain Category':'Medical Domain Category'},
             height=600,
             color_discrete_sequence=['#89c2d9']
)  

fig.update_layout(
    xaxis_title='',  
    yaxis_title='', 
    yaxis=dict(title_font=dict(size=16, family='Arial', color='black'), title_standoff=5),
    font=dict(size=14, family='Arial', color='black', weight='bold'),  
    plot_bgcolor='white',  
    margin=dict(l=10, r=10, t=50, b=10),
    width=1000,
)

fig.update_xaxes(linecolor='black')  
fig.update_yaxes(linecolor='black')
fig.update_traces(marker=dict(line=dict(color='#ddbea9', width=0.5))) 

fig.show()
```

```python
# picture of country
fig = go.Figure()

fig.add_trace(go.Choropleth(
    locations=['United States'],
    locationmode='country names',
    z=[1],  
    colorscale=[[0, '#89c2d9'], [1, '#89c2d9']],  
    zmin=0,
    zmax=1,
    showscale=False, 
))

fig.update_geos(fitbounds="locations", visible=True)

fig.update_geos(
    visible=False,  
    showcoastlines=False,
    showcountries=False,
    countrycolor='black',
)

fig.add_annotation(
    text='United States',
    xref='paper',
    yref='paper',
    x=0.75, 
    y=0.43,  
    showarrow=False,
    font=dict(size=25, color='black', family='Arial', weight='bold')
)

fig.update_layout(
    width=700,
    margin=dict(l=10, r=10, t=10, b=10)
)

fig.show()
```
